```python

# option 1

# Create a map object
m = folium.Map(
    location=[-15.335972, -47.915726],
    zoom_start=4,
    tiles='Stamen Terrain'
)

# Configure geojson layer
folium.GeoJson(
    geo_json_data,
    name='DistribuiÃ§Ã£o dos candidatos',
    style_function=lambda feature: {
        'fillColor': colormap(uf_dict[feature['properties']['uf_05']]),
        'color': 'black',
        'weight': 1,
        'dashArray': '5, 5',
        'fillOpacity': 0.9,
    }
).add_to(m)

colormap.caption = 'DistribuiÃ§Ã£o dos candidatos (2013)'
colormap.add_to(m)

folium.LayerControl().add_to(m)

m

```
