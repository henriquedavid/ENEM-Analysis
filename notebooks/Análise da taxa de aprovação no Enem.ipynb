```
import pandas as pd 

escola = pd.read_excel('PLANILHA_ENEM_ESCOLA_2014.xlsx', skiprows=[0])
rn_state = escola[escola['SIGLA DA UF'] == 'RN']
rn_state.head()
```

```
escolas_rn = {}
numero_escola = {}
cidade = []
for _, row in rn_state.iterrows():
    if row['MUNICÍPIO'] in escolas_rn:
        escolas_rn[row['MUNICÍPIO']] += row['TAXA DE APROVAÇÃO']
        numero_escola[row['MUNICÍPIO']] += 1
    else:
        escolas_rn[row['MUNICÍPIO']] = row['TAXA DE APROVAÇÃO']
        numero_escola[row['MUNICÍPIO']] = 1
        cidade.append(row["MUNICÍPIO"])

for r in cidade:
    escolas_rn[r] = escolas_rn[r]/numero_escola[r]
print(escolas_rn)
```

```
df = pd.DataFrame.from_dict(escolas_rn, orient='index')
df.reset_index(level=0, inplace=True)
df.columns = ['municipios','aprovacao']
df.head()
```

```
import os
import folium
import json
from branca.colormap import linear
import numpy as np
from numpy import random
import unidecode
```

```
geo_json_data = json.load(open('geojs-24-mun.json',encoding='latin-1'))
```

```
colormap = linear.RdPu_03.scale(
    df.aprovacao.min(),
    df.aprovacao.max())
```

```
for i in geo_json_data['features']:
  if not i['properties']['name'].upper() in escolas_rn:
    escolas_rn[i['properties']['name'].upper()] = 0
    
```

```

lat = [-5.663889, -6.4636721, -6.3794156, -5.6455161, -6.2420251, -5.5444205, -5.5343439, -5.6665552,-5.2315424,-5.1951504,-5.8350506, -6.4598912, -6.6916059, -5.9225221, -6.1027702, -6.2309344, -5.7747709, -5.9343692]
lon = [-37.798889, -37.104435, -35.1322001, -35.4419311, -36.5124361, -36.871496, -35.8218385, -36.3137277, -36.6999278, -37.3766595,-35.2069457, -35.510892, -36.6696983, -35.2461203, -38.2283202, -36.0313754, -35.3945459, -35.829152]
name = ['Apodi', 'Caicó', 'Canguaretama', 'Ceará Mirim', 'Currais Novos', 'Ipanguaçu', 'João Camara', 'Lajes', 'Macau', 'Mossoró', 'Natal', 'Nova Cruz', 'Parelhas', 'Parnamirim', 'Pau dos ferros', 'Santa Cruz', 'São Gonçalo do Amarante', 'São Paulo do Potengi']
```

```
m = folium.Map(
    location=[-5.66556, -36.60111],
    zoom_start=8,
    tiles='Stamen Terrain'
)

folium.GeoJson(
    geo_json_data,
    name='Média de aprovação nas cidades do RN',
    style_function=lambda feature: {
        'fillColor': colormap(escolas_rn[unidecode.unidecode(feature['properties']['name'].upper())]),
        'color': 'black',
        'weight': 1,
        'dashArray': '5, 5',
        'fillOpacity': 0.9,
    }
).add_to(m)


for i in range(18):
  folium.Marker(location=[lat[i], lon[i]], popup=name[i]).add_to(m)


colormap.caption = 'Taxa de aprovação no ENEM nas cidades do RN'
colormap.add_to(m)


m
```
