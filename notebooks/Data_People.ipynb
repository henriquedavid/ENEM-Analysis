```python
idd_2012 = data_2012.IDADE
idd_2012 = pd.DataFrame(idd_2012.value_counts())
idd_2012 = idd_2012.reset_index()

idd_2014 = data_2014.IDADE
idd_2014 = pd.DataFrame(idd_2014.value_counts())
idd_2014 = idd_2014.reset_index()
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

trace1 = go.Bar(
    x=idd_2012['index'],
    y=idd_2012['IDADE'],
    name='2012'
)
trace2 = go.Bar(
    x=idd_2014['index'],
    y=idd_2014['IDADE'],
    name='2014'
)

data = [trace1, trace2]
layout = go.Layout(
    barmode='group'
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='grouped-bar')
```

```python
data_2014_UF = data_2012.UF_INSC.value_counts()
data_2014_UF = pd.DataFrame(data_2014_UF).reset_index()
data_2014_UF.head()
```

```python
data_2014_UF.UF_INSC = pd.to_numeric(data_2014_UF.UF_INSC)
data_2014_UF
```

```python
# colormap yellow and green (YlGn)
colormap = linear.Blues_03.scale(
    data_2014_UF.UF_INSC.min(),
    (data_2014_UF.UF_INSC.max()))

print(colormap(590.0))

colormap
```

```python
uf_dict = data_2014_UF.set_index(data_2014_UF['index'])['UF_INSC']
uf_dict[:3]
```

```python

# option 1

# Create a map object
m = folium.Map(
    location=[-5.826592, -35.212558],
    zoom_start=7,
    tiles='Stamen Terrain'
)

# Configure geojson layer
folium.GeoJson(
    geo_json_data,
    name='Population estimation of RN State in 2017',
    style_function=lambda feature: {
        'fillColor': colormap(uf_dict[feature['properties']['uf_05']]),
        'color': 'black',
        'weight': 1,
        'dashArray': '5, 5',
        'fillOpacity': 0.9,
    }
).add_to(m)

colormap.caption = 'Population estimation (2017)'
colormap.add_to(m)

folium.LayerControl().add_to(m)

m

```
